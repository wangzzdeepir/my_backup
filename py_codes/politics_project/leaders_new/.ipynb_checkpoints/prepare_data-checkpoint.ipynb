```python
#!/usr/bin/python
"""This cell creates .txt files for convert_imageset in Caffe.
"""

img_root = '/home/deep/Documents/zongzuo/pythoncodes/leaders_new/leaders_images/images/'
with open('/home/deep/Documents/zongzuo/pythoncodes/leaders_new/leaders_images/annotations.txt') as txt:
    for line in txt:
        line = line.strip()
        if line.find('.')>=0:
            linebuf = line
        else:
            print img_root + ' '.join([linebuf, str(int(line)-1)])
            with open('data_list.txt', 'a') as txto:
                txto.write(img_root + ' '.join([linebuf, str(int(line)-1)])+'\n')
print 'Done.'
            
```

```python
#!/usr/bin/python
'''This cell splits datalist.txt into train_list and val_list.
'''

with open('/home/deep/Documents/zongzuo/pythoncodes/leaders_new/leaders_images/data_list.txt') as txt:
    for i, line in enumerate(txt):
        line = line.strip()
        if i%4 > 0:
            with open('train_list.txt','a') as train_list:
                train_list.write(line + '\n')
        else:
            with open('val_list.txt','a') as val_list:
                val_list.write(line + '\n')
print 'Done.'
```
