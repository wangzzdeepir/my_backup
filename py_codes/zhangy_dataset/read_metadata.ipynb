```python
#!/usr/bin/python
"""
"""
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

count = np.zeros((4790, 1)).astype('int32')
c = 0
rang_lst = np.arange(50, 150, 10)
c_rang = np.zeros((1, len(rang_lst)))
with open('/home/deep/Documents/zongzuo/py_codes/zhangy_dataset/readme.txt') as metadata:
    for line in metadata:
        line = line.strip()
        if line.find('全部') >= 0:
            number = int(line.partition('全部')[2].partition('张')[0])
            count[c] = number
            c += 1
            for c_i, rang in enumerate(rang_lst):
                if number >= rang:
                    c_rang[0, c_i] += 1
                

plt.hist(count, bins=100, range=(0, 1000))
print sum(count)
print c_rang
print rang_lst
```
